```python

# Initial imports
import pandas as pd
import hvplot.pandas
from pathlib import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

**Methodology:**

1. **Preprocessing the Data for PCA:**
   - **Data Collection:** Gather data on various cryptocurrencies, including their market capitalization, trading volume, price, and other relevant features.
   - **Data Cleaning:** Handle missing values, remove duplicates, and normalize the data to ensure consistency and accuracy.
   - **Feature Selection:** Select relevant features that will be used for Principal Component Analysis (PCA).

```python
# Load the crypto_data.csv dataset.
path = "crypto_data.csv"
crypto_df =pd.read_csv(path, index_col=0)
print(crypto_df.shape)
crypto_df.head(10)
```

```python
# Keep all the cryptocurrencies that are being traded.
crypto_df2 = crypto_df[crypto_df["IsTrading"] == True]
print(crypto_df2.shape)
crypto_df2.head(10)
```

```python
# Remove the "IsTrading" column. 
crypto_df2= crypto_df2.drop(columns="IsTrading")
print(crypto_df2.shape)
crypto_df2.head(10)
```

```python
# Remove rows that have at least 1 null value.
crypto_df2 = crypto_df2.dropna()
print(crypto_df2.shape)
crypto_df2.head(10)
```

```python
# Keep the rows where coins are mined.
crypto_df2 = crypto_df2[crypto_df2["TotalCoinsMined"] > 0]
print(crypto_df2.shape)
crypto_df2.head(10)
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
crypto_names_df = pd.DataFrame(crypto_df2["CoinName"])
print(crypto_names_df.shape)
crypto_names_df.head()
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
crypto_df2= crypto_df2.drop(columns="CoinName")
print(crypto_df2.shape)
crypto_df2.head(10)
```

```python
# Use get_dummies() to create variables for text features.
crypto_df3 = pd.get_dummies(crypto_df2, columns = ["Algorithm", "ProofType"])
print(crypto_df.shape)
crypto_df3.head(10)
```

```python
# Standardize the data with StandardScaler().
crypto_scaled = StandardScaler().fit_transform(crypto_df3)
crypto_scaled[0:5]
```

2. **Reducing Data Dimensions Using PCA:**
   - **Principal Component Analysis (PCA):** Apply PCA to reduce the dimensionality of the dataset. This step helps in simplifying the data while retaining the most important information.
   - **Explained Variance:** Analyze the explained variance to determine the number of principal components to retain for further analysis.

```python
# Using PCA to reduce dimension to three principal components.
# Initialize PCA model
pca = PCA(n_components=3)

# Get three principal components for the data.
crypto_pca = pca.fit_transform(crypto_scaled)
crypto_pca
```

```python
# Create a DataFrame with the three principal components.
df_crypto_pca = pd.DataFrame(
    data = crypto_pca, columns = ["PC 1", "PC 2", "PC 3"],index = crypto_df2.index
)
print(df_crypto_pca.shape)
df_crypto_pca.head(10)
```

3. **Clustering Cryptocurrencies Using K-means:**
   - **K-means Clustering:** Use the K-means algorithm to cluster the cryptocurrencies based on the principal components obtained from PCA.
   - **Optimal Number of Clusters:** Determine the optimal number of clusters using methods such as the Elbow Method or Silhouette Score.
   - **Cluster Analysis:** Analyze the characteristics of each cluster to understand the grouping of cryptocurrencies.

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Find the best value for K
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_crypto_pca)
    inertia.append(km.inertia_)
```

```python
# Create an elbow curve to find the best value for K.
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

Running K-Means with `k=4`

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(df_crypto_pca)

# Predict clusters
predictions = model.predict(df_crypto_pca)
predictions
```

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.

# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
# pd.concat([df1, df3], join="inner")
# pd.concat([df1, df4], axis=1)
clustered_df = pd.concat([crypto_df2, df_crypto_pca], axis=1, join="inner")

#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies. 
# YOUR CODE HERE
clustered_df['CoinName'] = crypto_names_df.CoinName

#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
# YOUR CODE HERE
clustered_df['class'] = model.labels_

# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```
