### Project Overview:

The relationship between Alzheimer's disease (AD) and cholesterol levels remains a subject of debate, with conflicting findings across various studies. While early research suggested that elevated low-density lipoprotein (LDL) cholesterol might contribute to AD risk, more recent studies have produced mixed results. Similarly, the protective role of high-density lipoprotein (HDL) cholesterol in reducing AD risk is not yet fully understood.

In this project, exploratory data analysis (EDA) was conducted on a dataset of 2,149 patients from diverse demographics to examine the association between cholesterol levels and AD diagnosis. The analysis revealed that the mean cholesterol level for AD patients was 225.56 mg/dL, a borderline high level. However, statistical tests, including correlation analysis, t-tests, and ANOVA, did not establish a significant relationship between LDL cholesterol and AD diagnosis. Despite these findings, challenges arose in handling categorical variables within logistic regression models, highlighting the need for advanced machine learning techniques such as one-hot encoding.

Given the complexity of AD and its multifactorial nature, cholesterol alone may not be a decisive predictor of the disease. Future research should explore additional clinical markers, such as cerebrospinal fluid tests and amyloid or tau protein levels, to improve AD prediction and deepen our understanding of its underlying mechanisms.

```python
# Create a function to fin out outliers
def findOutliers_IQR(df):
    '''
    Find Outliers in the dataframe
       args: df (dataframe)  
    return:
        outliers (dataframe)
    '''
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR
    
    outliers = df[((df <= lower_bound) | (df >= upper_bound))]
    
    return outliers
```

First perform Shapiro-Wilk test to determine whether a sample data comes from a normally distributed population.

```python
# Shapiro Test on the dataframe
stat, p_value = scipy.stats.shapiro(filtered_df['CholesterolTotal'])
print(f"statistic: {stat}")
print(f"p-value: {p_value}")
print(f"=========================\n")

# Set significance level
alpha = 0.05

# p-value > alpha: We fail to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Then, calculate t-statistic to determine if there is a significant difference between the means of two groups.

```python
# Perform the t-test
t_statistic, p_value = stats.ttest_ind(pos_result['CholesterolTotal'], 
                                       neg_result['CholesterolTotal'])

print(f"T-statistic: {t_statistic}")
print(f"p-value: {p_value}")
print(f"=========================\n")

# Set significance level
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# Perform One-Way ANOVA
group1 = pos_result['CholesterolTotal']
group2 = neg_result['CholesterolTotal']

f_statistic, p_value = stats.f_oneway(group1, group2)
print(f"F-Statistic: {f_statistic}, P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the group means.")
```

### **`Summary: Alzheimer's Disease and Cholesterol`**

The intricate relationship between Alzheimer's disease (AD) and cholesterol levels has been a subject of extensive research. Cholesterol, a lipid essential for various bodily functions, exists in different forms: high-density lipoprotein (HDL), often termed "good" cholesterol, and low-density lipoprotein (LDL), commonly known as "bad" cholesterol.


Early studies primarily focused on LDL cholesterol, suggesting a potential link with AD. Elevated LDL levels were associated with increased risk factors for cardiovascular disease, which is often comorbid with AD. This led to the hypothesis that vascular damage caused by high LDL cholesterol might contribute to the development of AD.


However, subsequent research has painted a more complex picture. While some studies continue to support a link between high LDL cholesterol and AD risk, others have found conflicting results. The role of HDL cholesterol has also emerged as an area of interest. Some studies suggest that lower HDL levels might be associated with an increased risk of AD, potentially due to its protective effects against inflammation and oxidative stress. The assumption/hypothesis for this project is based on these findings- Alzheimer's disease tied to cholesterol level.


To explore the relationship between AD and cholesterol, I performed EDA on Alzheimerâ€™s data collected from 2149 patients. The dataset collected from different demography. The outcome of EDA indicates that the mean Cholesterol level for patients diagnosed with AD is 225.56mg/dL which is a borderline level high. Correlation analysis was also completed to evaluate if there is any relationship between the predictor- LDL Cholesterol and the response variables- AD diagnosis. The result, however, indicates the correlation is not statistically significant. This result also supported with the t-statistic and ANOVA test output. On both analyses, the project failed to reject the null hypothesis- Alzheimer's disease not tied to cholesterol level.


In the regression analysis, I tried to run the model using multiple predicator-response combination. I faced a challenge while I was dealing with multiple categorical variables, fitting them in the logistic regression model is a difficult task. I believe using machine learning one-hot encoding will solve such an issue. 

In conclusion, the relationship between Alzheimer's disease and cholesterol is multifaceted and not fully understood. While some studies point to a potential association, more research is needed to clarify the exact nature of this link and to identify potential therapeutic targets. In addition to the variables included in this project, other clinical tests like a cerebrospinal fluid test might help in future Alzheimer's data analysis. Amyloid and tau proteins can be also another key predictor in the prediction of AD.
